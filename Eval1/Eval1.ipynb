```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mplt
import pylab as plt
from pylab import figure, show, legend, xlabel, ylabel
import statsmodels.api as sm
```

```python
Annual=pd.read_csv("Datos.csv", header=int(1))
```

```python
Annual
```

```python
Annual.columns 
```

```python
Annual_1= Annual[:-1]
```

```python
Annual.drop(['Jan','Mar','Feb','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec','D-N','DJF','MAM','JJA','SON'],1)
```

```python
x=Annual_1[u'Year']
y=Annual_1[u'J-D']
```

```python
mplt.plot(x,y,'ks')
mplt.plot(x,y,'k-')
mplt.grid(False)
plt.xlabel('AÃ±o')
plt.ylabel('AnomalÃ­as en temperatura')
plt.title('Datos mensuales combinados de superficie terrestre y superficie de mar')
```

```python
plt.show()
```

```python
x=Annual_1[u'Year']
y=Annual_1[u'J-D']
lowess = sm.nonparametric.lowess(y, x, frac=0.1)

plt.plot(lowess[:, 0], lowess[:, 1],'r-')
```

```python
plt.show()
```
