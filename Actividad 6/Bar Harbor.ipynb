```python
N = 1440
T = 1
x = df['Date Time']
y = df['Water Level']
yf = fft(y)
xf = fftfreq(N, T)
xf = fftshift(xf)
yplot = fftshift(yf)
plt.plot(xf, 1.0/N * np.abs(yplot))
plt.xlim(0.000000,0.1)
plt.xlabel('Frecuencia')
plt.ylabel('Water Level(m)')
plt.text(0.08, 0.85,'$M_2$')
plt.text(0.082, 0.2,'$L_2$')
plt.text(0.075, 0.2,'$N_2$') #0.0785
plt.text(0.0415, 0.1,'$S_2$')
plt.text(0.0367, 0.1,'$O_1$') #0.0385
plt.grid()
plt.show()
```
